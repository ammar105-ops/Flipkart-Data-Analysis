### Removing and changing the unwanted strings from dataset

```python
df['Storage'].value_counts()
```

```python
num_col_df = df.loc[:, num_attr] #Removing GB
for col in list(num_col_df):
    c = 0
    for i in num_col_df.loc[:,col]:
        if isinstance(i, str):
            if ('GB' in i):
                num_col_df.loc[c, col] = i.replace('GB', '')
        c = c+1
MBS = []     #Removing MB
MBM = []
for col in list(num_col_df):
    c = 0
    for i in num_col_df.loc[:,col]:
        if isinstance(i, str):
            if ('MB' in i and col=='Memory'):
                num_col_df.loc[c, col] = i.replace('MB', '')
                MBM.append(c)
            elif ('MB' in i and col=="Storage"):
                num_col_df.loc[c, col] = i.replace('MB', '')
                MBS.append(c)
        c = c+1
        
TBS = []     #Removing TB
TBM = []
for col in list(num_col_df):
    c = 0
    for i in num_col_df.loc[:,col]:
        if isinstance(i, str):
            if ('TB' in i and col=='Memory'):
                num_col_df.loc[c, col] = i.replace('TB', '')
                TBM.append(c)
            elif ('TB' in i and col=="Storage"):
                num_col_df.loc[c, col] = i.replace('TB', '')
                TBS.append(c)
        c = c+1

for col in list(num_col_df):
    c = 0
    for i in num_col_df.loc[:,col]:
        if isinstance(i, str):
            if ('Expandable Upto 32' in i):
                num_col_df.loc[c, col] = i.replace('Expandable Upto 32', '32')
        c = c+1
        
for col in list(num_col_df):
    c = 0
    for i in num_col_df.loc[:,col]:
        if isinstance(i, str):
            if ('Expandable Upto 16' in i):
                num_col_df.loc[c, col] = i.replace('Expandable Upto 16', '16')
        c = c+1
        
model_df = df.loc[:, 'Model'] #Removing blank spaces after model names
c = 0
for i in model_df:
    model_df.loc[c] = i.rstrip()
    c = c+1
df['Model'] = model_df
```

### Ques.9 Predict the rating of an Apple Phone

```python
df1 = df[df["Brand"]== 'Apple']
df1.head()
```

```python
df2 = df1.drop("Brand", axis = 1)
```

```python
df1["Model"].value_counts()
```

```python
df1["Color"].value_counts()
```

##### Splitting the dataset

```python
x = df2.drop('Rating', axis = 1)
y = df2[['Rating']]
```

##### Applying transformations

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OrdinalEncoder
attr = ["Model", "Color"]
encoder = OrdinalEncoder()
x.loc[:, attr] = encoder.fit_transform(x.loc[:, attr])
```

```python
encoder.categories_
```

```python
from sklearn.decomposition import PCA
pca = PCA(n_components=0.95)
x2 = pca.fit_transform(x)
```

```python
pca.explained_variance_ratio_
```

##### Training the model

```python
from sklearn.linear_model import LinearRegression

regressor = LinearRegression()
regressor.fit(x2, y)
```

##### Taking the inputs from user and processing it

```python
model = input(print("Enter Model of the iPhone: "))
color = input(print("Enter colour of the iPhone: "))
memory = float(input(print("Enter RAM memory of the iPhone in GB: ")))
storage = float(input(print("Enter storage of the iPhone in GB: ")))
OP = float(input(print("Enter original price of the iPhone: ")))
SP = float(input(print("Enter discounted/selling price of the iPhone: ")))
```

```python
full_attr = list(x)
test = pd.DataFrame(columns = full_attr)
values = [model, color, memory, storage, OP, SP]
D = dict()
for i, j in zip(full_attr, values):
    D[i] = [j]
df = pd.DataFrame(D)
test = test.append(df, ignore_index = True)

test.loc[:, attr] = encoder.transform(test.loc[:, attr])
test1 = pca.transform(test)
```

##### Printing the results

```python
print(f'The configurations of the iPhone is given by:\n 1. Model - {model} \n 2. Color - {color} \n 3. Memory - {memory}GB RAM \n 4. Storage - {storage}GB \n 5. Original Price - {OP} \n 6. Discounted/Selling Price - {SP}')
pred = regressor.predict(test1)
for i in pred:
    for j in i:
        print("The predicted rating for the iPhone is = ", j)
```

### Ques.10 Calculate the errors in the prediction

```python
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error

pred = regressor.predict(x2)
mse = mean_squared_error(y, pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y, pred)
```

```python
print("The mean squared error in predictions = ", mse)
print("The root mean squared error in predictions = ", rmse)
print("The mean absolute error in predictions = ", mae)
```
